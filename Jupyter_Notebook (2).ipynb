```python
import yfinance as yf

# Extraction des données boursières de Tesla (TSLA)
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="5y")

# Affichage des cinq premières lignes
tesla_data.head()
```

```python
# Extraction des données boursières de GameStop (GME)
gamestop = yf.Ticker("GME")
gamestop_data = gamestop.history(period="5y")

# Affichage des cinq premières lignes
gamestop_data.head()
```

```python
import plotly.graph_objs as go
from plotly.subplots import make_subplots

# Création du graphique des prix des actions de Tesla
fig_tesla = make_subplots(rows=2, cols=1, shared_xaxes=True, vertical_spacing=0.1)

# Tracé des prix des actions de Tesla
fig_tesla.add_trace(go.Scatter(x=tesla_data.index, y=tesla_data['Close'], name='Tesla Stock Price'), row=1, col=1)

# Tracé des revenus de Tesla
dates = [entry[0] for entry in tesla_revenue]
revenues = [float(entry[1].replace('$', '').replace('B', '').replace('M', '')) * (10**9 if 'B' in entry[1] else 10**6) for entry in tesla_revenue]
fig_tesla.add_trace(go.Bar(x=dates, y=revenues, name='Tesla Revenue'), row=2, col=1)

# Mise en forme du graphique
fig_tesla.update_layout(title='Tesla Stock Price and Revenue', height=600)
fig_tesla.show()
```
